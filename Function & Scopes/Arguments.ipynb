# Function Argument

An argument is a value that is passed to a function when it is called. It might be a variable, value or object passed to a function or method as input.

There are four types of arguments that we can provide in a function:

1) Default Argument
2) Positional Argument
3) Keyword Argument
4) Variable length Argument
5) Required Argument

Positional Arguments are needed to be included in proper order i.e the first argument is always listed first when the function is called, second argument needs to be called second and so on, values are assigned to parameters based on their position.

Values are matched by order when calling a function.

```python
def fun(First_name,Last_name):
    result=First_name+" "+Last_name
    print(result)
fun("Sagar","Singh")
```

Default Arguments in this we can provide a default value while creating a function.This way the function assumes a default value even if a value is not provided in the function call for that argument.

When defining a function, if we assign a value to a parameter, it becomes optional while calling.

```python
def fun(a=10,b=20):
    return (f"The sum of {a} and {b} is {a+b}")
    print()
```

```python
fun()
```

```python
fun(20,30)
```

```python
fun(a=20,)
```

```python
fun(b=100)
```

```python
fun(a=20)
```

Keyword Arguments we can provide arguments with key=value, this way the interpreter recognizes the arguments by the parameter name. Hence, the order in which the argument are passed does not matter.

Values are passed as key=value, so order doesn’t matter.

```python
def fun(First_name, Last_name):
    print(First_name, Last_name)

fun(Last_name="Singh", First_name="Sagar")   # Sagar Singh
```

Required Arguments When defining a function, if we do not assign a default value, that parameter is compulsory to pass while calling.

```python
def greet(first, last):
    print("Hello" ,first ,last)

greet("Sagar", "Singh")
```

## Variable lenght argument

### There are two types:

```python
def avg(*numbers):
    sum=0
    for i in numbers:
        sum=sum+i
    print("The average is:",sum/len(numbers))

avg(10,20,30,40)
```

```python
def avg(*numbers):
    user_input=(input("Enter the input").split())
    sum=0
    for i in int(user_input):
        sum=sum+i
    print("The average is:",sum/len(numbers))

avg(10,20,30,40)
```

```python
def add(*numbers):
    return sum(numbers)
    print()
```

```python
add(1,48,5647,47476,774)
```
