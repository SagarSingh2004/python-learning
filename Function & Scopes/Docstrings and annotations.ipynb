# Docstrings

```python
def add(a, b):
    """
    This function takes two numbers as input
    and returns their sum.
    """
    return a + b
```

```python
print(add.__doc__)
```

# Annotations

```python
def function(a:str,b:str)->str:
    """My name is sagar"""
    return a+b
```

```python
function(1,2)        
```

```python
print(function.__annotations__)
```
