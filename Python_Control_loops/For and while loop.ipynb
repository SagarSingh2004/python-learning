# Loops

### for loop

A for loop in python is used to repeat a block of code a specific number of times or iterate over a collection( like a list, string, or range).

```python
#for variable in sequence:
    #code block
```

1) variable: takes each value from the sequence one by one.
2) sequence: can be a list, tuple, string, range, etc.

```python
fruits=["apple","mango","banana"]
for items in fruits:
        print(items)
```

```python
a="Sagar Singh"
for char in a:
    print(char)
```

```python
for r in range(0,10):
    print(r)
```

```python
for r in range(10,1,-1):
    print(r)
```

```python
for r in range(50,1,-5):
    print(r)
```

```python
for r in range(1,100+1):
    print(r)
```

```python
list=["Sagar","Varsha","Papa","Mummy"]
for items in list:
    print(items)
    for i in items:
        print(i)
```

```python
for i in range(1,10):
    print(i)
```

```python
for i in range(2,21,2):
    print(i)
```

### while loop

A while loop repeatedly executes a block of code as long as the condition is True.

It's used when we don't know in advance how many times you need to repeat something(unlike for loops which usually iterate over a sequence or range).

```python
#while condition:
    # code to execute repeatedly
```

1) condition: a boolean expression (True or False)
2) the code inside the loop runs as long as the condition is True.
3) Once the condition becomes False, the loop stops.

```python
i=1
while i<=5:
    print(i)
    i+=1
```

```python
password=""
while password != "1234":
    password=input("Enter password:")
print("Access Granted")
```

```python
i=0
while i<3:
    print(i)
    i=i+1
```

```python
i=""
while i!="Sagar":
    i=input("What is your name")
print("Done! Access Granted")
```

```python
i=1
while i<=38:
    i=int(input("What is i"))
print("Done")
```

```python
i=10
while i>=0:
    print(i)
    i-=1
```

### Else with while loop

```python
i=1
while i>=15:
    print(i)
    i=i-1
else:
    print("The loop is not started")
```
