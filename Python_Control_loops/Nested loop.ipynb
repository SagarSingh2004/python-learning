# Nested Loops

The loop inside another loop is known as nested loop.

outer loop first iteration runs then inner loop runs all its iteration

Why use nested loops?
1) Handling Multi-Dimensional Data: Such as matrices, grids, or lists of lists.
2) Complex Iterations: Operations depend on multiple variables or dimensions.
3) Pattern Generation: Creating patterns, such as in graphics or games.

```python
for i in range(1,4):
    print(i)
```

```python
for i in range(0,3):
    for j in range(0,3):
        print(j)
    print()
```

```python
i=1
while i<4:
    for j in range(1,4):
        print(j)
    i+=1
```

```python
i=1
while i<=5:
    for j in range(0,i):
        print("*",end=" ")
    print()
    i+=1
```

```python
i = 1                  # start row
rows = 5               # total rows

while i <= rows:        # outer loop → while
    for j in range(i):  # inner loop → for
        print("*", end=" ")
    print()            # new line after each row
    i += 1             # increment row
```
