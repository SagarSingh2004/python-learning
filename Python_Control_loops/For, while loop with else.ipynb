# For loop with else:

The meaning of else content print is basically the loop ends if loop break than else doesn't print.

```python
for i in range(5):
    print(i)
else:
    print("sorry")
```

```python
i=5
while i>=0:
    print(i)
    i-=1
else:
    print("Hello")
```

```python
for i in range(5):
    print(i)
    if i==4:
        break
else:
    print("sorry")    # like that it break when i is 4 than else doesn't print
```
