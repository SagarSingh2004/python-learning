# Introduction to python

### What is python

1) python is a high-level, interpreted, general-purpose programming language.
2) Created by Guido van Rossum in 1991.
3) Known for its simplicity, readability, and versatility.

### Features of python

1) Simple & easy to learn- syntax is like english language.
2) Interpreted- runs line by line.
3) Portable- runs on windows, Linux, Mac.
4) Object oriented- supports classes & objects.
5) Extensible- can integrate with c, c++, java.
6) Open source programming language.
7) Huge libraries- Numpy, pandas, Tensorflow, etc.

### Python Applications

1) Web development- Django, flask.
2) Data Science- Numpy, Pandas.
3) ML- scikit-learn.
4) AI- Tensorflow.
5) Game development- Pygame.
6) Desktop Apps- Tkinter.
7) IoT & Robotics- Raspberry Pi
