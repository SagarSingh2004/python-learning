# Variables And Data types

## Variable

1) A variable is a name that stores a value in memory.
2) In python don't need to declare type explicitly.

### Rules for naming Variables

1) Must start with a letter or _.
2) Cannot start with a number.
3) No special characters ( @, $, %).
4) Case sensitive ( Age and age are different variables).

## Data Types

Python has built in data types
1) Interger(int)- whole numbers, i.e. 1,2,3,etc.
2) Floating point numbers(float)- Decimal numbers, i.e. 1.0,2.0,etc.
3) Complex numbers(complex)- Numbers with real and imaginary part, i.e. 3+4j.
4) String(str)- sequence of characters, i.e. "Hello". Strings are always enclosed in a single or double quote.
5) List(list)- ordered, mutable collection of items, i.e. [1,2,3]. Lists are denoted by square brackets[].
6) Tuples(tuple)- ordered, unmutable collection of items, i.e. (1,2,3),("a","b","c"). Tuples are denoted by parantheses().
7) Range(range)- a sequence of numbers, i.e. range(1,10).
8) Dictionaries(dict)- ordered collections of key value pairs, i.e. {"name":"sagar","age":21}. Dict are denoted by curly brackets{}.
9) Set- unique collection, i.e. {1,2,3}.
10) Boolean(bool)- A logical value that can be either True or false.
11) None(None type)- A special value that represents the absence of a value.
12) frozenset: same as sets but immutable,Useful when you need a set that stays fixed, i.e. frozenset({1,2,3}).
