### Reverse

Reverse the whole list.

```python
l1=["Sagar",1,4,1.5,True]
l1.reverse()
print(l1)
```

### index

This method returns the index of the first occurrence of the list item.

```python
l1=["Sagar",1,4,1.5,1,True]
print(l1.index(1))
print(l1.index(1,3))
```

### Count

this method returns the count of the number of items with the given value.

```python
l1=["Sagar",1,4,1,False]                 #True is als count as 1, false is count as 0
print(l1.count(1))
```

### Copy

Returns copy of the list. This can be done to perform operations on the list without modifying the original list.

```python
l=[i for i in range(10)]
print(l)
```

```python
m=l                 
m[0]=10                #This also change the item in original list l.
print(l)
```

```python
m=l.copy()                # This just copy the list.
m[0]=0   
print(l)
```

```python
print(m)
```

### Insert

This method inserts an item at the given index position.User has to specify index and the item to be inserted within the insert() method.

```python
l1=["Sagar","Singh","Rawat"]
l1.insert(1,"Cast")
print(l1)
```

### Extend

This methods adds an entire list or any other collection data type (set, tuple, dict) to the existing list.

```python
l=[1,2,3,4,5,6,7,8,9,0]
m=(1,2,3,4,5,6,7,8,9,0)
l.extend(m)
```

```python
print(l)
```

### Concatenating new lists:
