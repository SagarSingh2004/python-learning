# Set

An unorderd collection of unique elements.

```python
# You can define a set using curly braces
s={2,4,6,8,10,"A","a"}
print(type(s))
print(s)
```

```python
# An empty set must be created using set(), not{} because curly braces are used for empty dictionary
a={}
b=set()
type(a),type(b)
```

### Properties of sets

1) Unordered: No index based access( cannot do s[0]).
2) Mutable: you can add/remove items.
3) Unique items: Duplicates are automatically removed.

```python
Name={"Sagar","Ashutosh","Krishna","Jatin","Sagar"}
print(Name)
```

```python
for nam in Name:
    print(nam,end=" ")
```

# Sets Methods

# Union (|)

a set union combines elements from two (or more) sets, keeping only unique elements (no duplicates), returns a new set without changing the originals.

```python
s1={1,2,3,4}
s2={4,5,6}
s3=s1.union(s2)
print(s3)
```

# Update

This method modifies the original set in-place.

```python
s1={1,2,3,4}
s2={4,5,6}
s1.update(s2)
print(s1)
```

# Intersection

a set intersection prints only items that are similar in both sets.

```python
s1={x for x in range(60) if x%3==0}
s2={x for x in range(60) if x%6==0}
s3=s1.intersection(s2)
print(s3)
```

# Intersection_update

This methods modifies the original set in place.

```python
s1={x for x in range(60) if x%3==0}
s2={x for x in range(60) if x%6==0}
s1.intersection_update(s2)
print(s1)
```

# symmetric_difference

```python
s1 = {1, 2, 3, 4}
s2 = {3, 4, 5, 6}
result = s1.symmetric_difference(s2)
print("symmetric_difference:", result)   # {1, 2, 5, 6}
print("s1 after:", s1)                   # {1, 2, 3, 4} (unchanged)
```

# symmetric_difference_update()

```python
s1 = {1, 2, 3, 4}
s2 = {3, 4, 5, 6}
s1.symmetric_difference_update(s2)
print("s1 after update:", s1)   # {1, 2, 5, 6}
```

# difference()

```python
s1 = {1, 2, 3, 4}
s2 = {3, 4, 5, 6}
result = s1.difference(s2)
print("difference:", result)   # {1, 2}
print("s1 after:", s1)         # {1, 2, 3, 4} (unchanged)
```

# difference_update()

Updates the original set in place, removing elements found in the other set(s).

```python
s1 = {1, 2, 3, 4}
s2 = {3, 4, 5, 6}
s1.difference_update(s2)
print("s1 after update:", s1)   # {1, 2}
```

# isdisjoint() #No same element in two sets

This method checks if items of given sets are in another set. it returns false if items are present else True.

```python
s1={1,2,3}
s2={4,5,6}
print(s1.isdisjoint(s2))
```

# issuperset() #all the items are in another set

if a set have all the element of another set than the set is said to be superset. A set A is a superset of set B if all elements of B are in A.

```python
s1={3,6,9,12,15,18,21,24,27,30}
s2={6,12,18,24,30}
print(s1.issuperset(s2))
```

# issubset()

A set A is a subset of set B if all elements of A are in B.

```python
s1={3,6,9,12,15,18,21,24,27,30}
s2={6,12,18,24,30}
print(s1.issubset(s2))
```

# Remove() or discard()

Both remove items from a set but remove raises an error if the item is not present in the cell and discard not raises error eventhough the items is not present.

```python
s1={1,2,3,4}
s1.remove(4)
print(s1)
```

```python
s1={1,2,3,4}
s1.remove(5)
```

```python
s1={1,2,3,4}
s1.discard(4)
print(s1)
```

```python
s1={1,2,3,4}
s1.discard(5)
```

# pop

This method removes the last item of the set but the catch is we don't know which item get popped as sets are unordered. however, you can access the popped item if you assignthe pop method to a variable.

```python
s1={1,2,3,4}
items=s1.pop()
print(s1)
print(items)
```
