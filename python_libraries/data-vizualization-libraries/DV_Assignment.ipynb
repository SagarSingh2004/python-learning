```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("indian_food.csv")
```

```python
df.sample(10)
```

```python
#replace all the"-1" of state column to the mode of state column
most_common_state=df["state"].mode()[0]
df["state"]=df["state"].replace("-1",most_common_state)
```

```python
#replace all the "-1" of region column to the region of the mode of state
most_common_region=df.loc[(df["state"]==most_common_state) & (df["region"]!="-1"),"region"].mode()[0]
df["region"]=df["region"].replace("-1",most_common_region)
```

```python
#replace the -1 with mode of the cook time
avg_cook_time=df.loc[df["cook_time"]!=-1,"cook_time"].mean()
df["cook_time"]=df["cook_time"].replace(-1,avg_cook_time).astype(np.int64)
```

```python
df.loc[(df["state"]=="-1") | (df["region"]=="-1") | (df["cook_time"]==-1)]      #To check that we replace -1 successfully for all
```

### Pie chart - Customer prefered flavor profile

```python
flavors=df["flavor_profile"].value_counts()

plt.figure(figsize=(6,6),dpi=100)
plt.pie(flavors.values,labels=flavors.index,autopct="%1.1f%%",colors=sns.color_palette("pastel"),shadow=True)
plt.title("Customer-Preferred Flavor Profiles")
plt.show()
```

### Bar Chart-Most requested courses

```python
course=df["course"].value_counts().sort_values(ascending=False)

plt.figure(figsize=(6,6),dpi=100)
bars=plt.bar(course.index,course.values,color=sns.color_palette("pastel"),edgecolor="k")
plt.title("Most requested course by customers")
for bar in bars:
    height=bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,
        height,
        str(height),
        ha="center",
        va="bottom",
        fontsize=10)
plt.xlabel("course")
plt.ylabel("customer_count")
plt.show()
```

### Grouped Bar Chart-diet order by region

```python
diet=df.groupby(["region","Diet"]).size().unstack()

plt.figure(figsize=(5,5),dpi=100)
width=0.4
x=np.arange(len(diet.index))

bars1=plt.bar(x+width/2,diet["non vegetarian"],width=width,color="green",label="Non-Vegetarian",edgecolor="k")
bars2=plt.bar(x-width/2,diet["vegetarian"],width=width,color="red",label="Vegetarian",edgecolor="k")    
plt.xticks(x,diet.index,rotation=45)
plt.title("Non_Veg_Vs_Veg_diet_order_by_region")
plt.xlabel("Region")
plt.ylabel("Customer_counts")
plt.legend()
plt.show()
```

### Box plot-order cooking time

```python
cook_time=df["cook_time"].dropna()

plt.figure(figsize=(10,10),dpi=100)
plt.boxplot(cook_time,vert=1,patch_artist=True,showfliers=1)
plt.title("Cook_time_distribution")
plt.ylabel("Cook_time")
plt.grid(True,linestyle="-.")
plt.show()
```

5 types of charts i am familiar with are

1. Line chart
2. Bar chart
3. scatter chart
4. pie chart
5. histogram

### Line chart

```python
x=np.arange(0,10)
y=x**2

plt.figure(figsize=(5,5),dpi=100)
plt.plot(x,y,"r-.o")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Line chart")
plt.grid(True,linestyle="-.")
plt.show()
```

### Bar chart

```python
x=np.arange(0,10)
y=x**2

plt.figure(figsize=(5,5),dpi=100)
plt.bar(x,y,color="r")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Bar chart")
plt.grid(True,linestyle="-.")
plt.show()
```

### Scatter Chart

```python
x=np.arange(0,10)
y=x**2

plt.figure(figsize=(5,5),dpi=100)
plt.scatter(x,y,color="r",s=100,marker="^")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Scatter chart")
plt.grid(True,linestyle="-.")
plt.show()
```

### Pie chart

```python
x=[10,15,40,100]
label=["A","B","C","D"]


plt.figure(figsize=(5,5),dpi=100)
plt.pie(x,labels=label,autopct="%1.1f%%",shadow=True,colors=sns.color_palette("pastel"),startangle=90)
plt.title("Pie chart")
plt.show()
```

### Histogram

```python
x=np.random.randint(1,100,50)

plt.hist(x,bins=10,edgecolor="k")
plt.xlabel("x")
plt.ylabel("Frequency")
plt.title("Histogram")
plt.grid(True,linestyle="-.")
plt.show()
```
