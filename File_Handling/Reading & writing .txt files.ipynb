| Mode   | Meaning            | Example                  |
| ------ | ------------------ | ------------------------ |
| `"r"`  | Read (default)     | `open("data.txt", "r")`  |
| `"w"`  | Write (overwrites) | `open("data.txt", "w")`  |
| `"a"`  | Append             | `open("data.txt", "a")`  |
| `"r+"` | Read & Write       | `open("data.txt", "r+")` |

### writing a file

```python
f=open("sagar.txt",'w')
f.write("My name is sagar.\n")
f.write("I am 21 years old")
f.close()
```

```python
f=open("sagar.txt","r")
lines=f.read(35)
print(lines)
f.close()
```

```python
with open("sagar.txt","w") as f:
    f.write("I live in faridabad")
```

### appending a file

```python
f=open("sagar.txt","a")
f.write(" Haryana")
```

```python
with open("sagar.txt","a") as f:
    f.write("\n My name is sagar")
```
