# Scope

```python
def func():
    x = 10  # local variable
    print(x)

func()
# print(x)  #Error, x is not accessible here
```

```python
def outer():
    y = 20
    def inner():
        print(y)  # can access outer function variable
    inner()

outer()
```

```python
z = 30  # global

def func():
    print(z)  # accessible inside function

func()
print(z)  # accessible outside too
```

```python
print(len("hello"))  # built-in functions
```

```python
a = 5  # global

def test():
    a = 10  # local
    print("Inside function:", a)

test()
print("Outside function:", a)
```

```python
count = 0

def increment():
    global count
    count += 1

increment()
print(count)  # 1
```

```python
def outer():
    x = 5
    def inner():
        nonlocal x
        x += 1
    inner()
    print(x)

outer()  # 6
```
