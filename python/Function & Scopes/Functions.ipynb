# Functions

A function is a block of code that performs a specific task, whenever it is called.

In bigger programs, where we have large amounts of code, it is advisable to create or use existing functions that make the program flow organized and neat.

There are two types of functions:
1) Built-in functions
2) User-defined functions

2) User-defined functions: We can create functions to perform specific tasks as per our needs. such functions are called user defined functions.

### Example: Function with parameters

```python
def add(a,b):
    result=a+b
    print(result)
```

### Example: Function with default parameter

```python
def Hello(name="Sagar"):
    result=(f"Hello {name}!")
    print(result)
```
