# Recursion

Recursion is when a function calls itself to solve a problem.

```python
def factorial(n):
    if n==0 or n==1:
        return 1
    else:
        return (n*factorial(n-1))
```

```python
factorial(3)
```

```python
def fibonacci_sequence(n):
    if n==0:
        return 0
    if n==1:
        return 1
    else:
        return fibonacci_sequence(n-1)+fibonacci_sequence(n-2)
```

```python
fibonacci_sequence(10)
```

```python
def sum(n):
    if n==0:
        return 0
    if n==1:
        return 1
    else:
        return n+sum(n-1)
```

```python
sum(5)
```

```python
def reverse(s):
    if s=="":
        return ""
    else:
        return reverse(s[1:])+s[0]
```

```python
reverse("sagar")
```

```python
def sum(x):
    if x==[]:
        return 0
    else:
        return x[0]+sum(x[1:])
    
```

```python
x=[1,2,3,4]
sum(x)
```
