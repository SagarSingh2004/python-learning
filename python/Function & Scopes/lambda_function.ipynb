# Lambda Functions

```python
# Syntax
```

lambda arguments: expression


```python
square=lambda x:x**2
print(square(10))
```

### Multiple arguments

```python
avg=lambda x,y: (x+y)/2
a=(avg(10,20))
print(a)
```

```python
def appl(a,b,c,d):
    return c+a(b,d)
c=appl(avg,10,10,10)
```

```python
print(c)
```

```python
add=lambda x:x+10
print(add(5))
```

```python
square=lambda x:x**2
print(square(8))
```

```python
product=lambda x,y:x*y
print(product(10,10))
```

```python
even=lambda x: True if x%2==0 else False
print(even(2))
```

```python
even=lambda x: "Truee" if x%2==0 else "Falsee"
print(even(2))
```

```python
abs=lambda x: x if x>0 else -x
print(abs(-0))
```

```python
ed=lambda x: "Zero" if x==0 else("positive" if x>0 else "Negative")
print(ed(10))
```

These are the higher order function means we run function as an argument in that

### Map()

```python
l=[1,2,3,4]
new_list=map(lambda x:x**3,l)
print(new_list)
```

```python
l=[1,2,3,4]
new_list=list(map(lambda x:x**3,l))
print(new_list)
```

### Filter()

```python
l=[1,2,3,4,5]
new_list=filter(lambda x:x>3,l)
print(new_list)
```

```python
l=[1,2,3,4,5]
new_list=list(filter(lambda x:x>3,l))
print(new_list)
```

### Reduce()

```python
from functools import reduce
```

```python
l=[1,2,3,4]
new_list=reduce(lambda x,y:x+y,l)
print(new_list)
```

```python
l=[11,22,88,99,55]
new_list=reduce(lambda x,y: x if x>y else y,l)
print(new_list)
```
