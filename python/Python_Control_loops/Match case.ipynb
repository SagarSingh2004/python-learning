# Match Case Statement

1) match check the values of a variable/ expresssion.
2) case compare it with patterns.
3) _ meaning match anything, it acts like a default case like else in if-else statement.

```python
a=1
match a:
    case 1:
        print("yes")
    case 1:
        print("yes 2")
    case _:
        print("NO")
```

```python
variable="Sagar"
match variable:
    case "Sagar":
        print("yes he is sagar")
    case _:
        print("No he is not sagar")
```

```python
x=int(input("Enter the value of x"))
match x:
    case 0:
        print("The number is zero")
    case 10:
        print("The number is positive integer")
    case _:
        print("The number is negative integer")
```

```python
x=int(input("Enter the value of x"))
match x:
    case 0:
        print("The number is zero")
    case 1:
        print("The number is 1")
    case _ if x<0:
        print(x,"- The number is negative integer")
    case _ if x>0:
        print(x,"- The number is Positive integer")
```

```python
Day=input("Enter the day").lower()
match Day.lower():
    case "saturday"|"sunday":
        print("Good it\'s a holiday")
    case "monday"|"tuesday"|"thursday"|"wednesday"|"friday":
        print("Oh it\'s a working day")
    case _:
        print("it\'s an invalid day")
```

```python
x=int(input("Enter your number"))
match x:
    case x if x==0:
        print("x is zero")
    case x if x>0:
        print("x is positive integer")
    case x if x<0:
        print("x is negative integer")
```
