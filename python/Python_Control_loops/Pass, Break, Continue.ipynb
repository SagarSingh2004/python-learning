# Break, Continue and pass

### break statement

The break statement enables to get out of the loop completely.

### continue statement

The continue statement skip just one round, but keep looping

### pass

It does nothing, it is just a placeholder where you want syntactically correct code but don't want it to do anything(yet).

```python
for i in range(0,5):
    if i==3:
       pass   # we add pass to make its syntax correct otherwise it don't work properly
    print(i)
```

```python
for i in range(12):
    print(i)
```

```python
for i in range(10):
    print("5 *",i+1,"=",5*(i+1))
```

```python
for i in range(10):
    print("5 *",i+1,"=",5*(i+1))
    if i==4:
        break
print("Escape Successfully")
```

```python
i=1
while i<=10:
    print("5 *",i,"=",5*i)
    i=i+1
    if i==6:
        break
print("escape successfully")
```

```python
for i in range(1,10+1):
    if i==5:
        continue
    print(i)
    
```

```python
for i in range(1,10+1):
    if i==5:
        print("Skip the 5*5")
        continue
    print("5 *",i,"=",5*i)
```

```python
i=0
while i<=10:
    i+=1
    if i==5:
        continue
    print("5 *",i,"=",5*i)
    
```
