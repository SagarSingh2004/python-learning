# List comprehension

List comprehension is a short and clean way to create a list in python.

```python
# General Syntax
#[expression for item in iterable if condition]
```

```python
lst=[i for i in range(4)]
```

```python
print(lst)
```

```python
lst=[i*i for i in range(4)]
```

```python
print(lst)
```

```python
lst=[i for i in range(4) if i%2==0]
print(lst)
```

### With if...else

```python
marks=[i if i%2==0 else -i for i in range(5)]
print(marks)
```

```python
marks=[i if i%3==0 else -i for i in range(10)]
print(marks)
```
