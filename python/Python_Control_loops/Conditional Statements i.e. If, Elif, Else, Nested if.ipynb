We work with conditional operators in it i.e. ==, !=, <, >, <=, >=

### if Statement

Used to execute a block of code only if a condition is True.

```python
a=10
if a>5:
    print("a is greater than 5")
```

### else Statement

Used to execute a block of code if the condition is false.

```python
a=10
if a>15:
    print("a is greater than 15")
else:
    print("a is less than 15")
```

```python
a=int(input("Enter Your Age:"))
print("Your age is",a)
if a>18:
    print("You can drive")
else:
    print("Your age is less than 18 so you can't drive")
```

```python
a=5
b=7
if not (a==5 and b==7):
    print("a and b is not equal to 5 and 7")
else:
    print("a and b is equal to 5 and 7")
```

```python
num=int(input("Enter the value of number"))
if num<0:
    print("The number is negative integer")
elif num==0:
    print("The number is zero")
else:
    print("The number is positive integer")
```

```python
x=int(input("Enter the number"))
if x==0:
    print("the number is zero")
elif x%2==0:
    print("the number is even")
else:
    print("the number is odd")
    
```

### elif Statement

Short for "else if". It's used to check multiple conditions.

```python
a=10
if a>10:
    print("a is greater than 10")
elif a<10:
    print("a is less than 10")
else:
    print("a is equal to 10")
```

```python
x=10
y=20
if x>15 and y>15:
    print("Both are greater than 15")
elif x>15 or y<15:
    print("x is greater and y is smaller than 15")
else:
    print("x is smaller and y is greater than 15")
```

### Important Notes

1) You can have only one if, zero or more elif, and zero or one else.
2) Conditions are evaluated top to bottom. The first True condition executes, and the rest are skipped.
3) You can use logical operators(and, or, not) to comine conditions.
4) All statements with the same level of indentation are considered part of the same block. Indentation is achieved using whitespace (spaces or tabs) at the beginning of each line.

### Nested if statement

If we can also place an if statement inside another if statement, then it is said to be nested if.

```python
num=int(input("Enter your number"))
if num<0:
    print("Your number is negative")
elif num>0:
    if num<=10:
        print("Number is between 1-10")
    elif num<=20 and num>10:
        print("Number is between 10-20")
    else:
        print("Number is greater than 20")
else:
    print("Number is zero")
```
