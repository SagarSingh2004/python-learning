### Checking for item in tuple

```python
country=("India","Pakistan","Bangladesh","China","Srilanka")
if "China" in country:
    print("Yes")
else:
    print("No")
```

### Manipulating tuple

first you have to convert tuple into list then you add or remove items then again convert into tuple.

```python
country=("India","Pakistan","Bangladesh","China","Srilanka")
lst=list(country)
lst.append("Russia")
lst.pop(1)
country=tuple(lst)
print(country)
```

### However we can concatenate two tuples

```python
tup1=("India","Bangladesh")
tup2=("China","Africa")
tup=tup1+tup2
print(tup)
```

# Tuple methods

### Count()

this methods returns the number of times the given element appears in the tuple.

```python
tup=(1,2,3,4,5,6,7,8,"Sagar")
tup2=tup*10
tup2.count("Sagar")
```

### Index()

This methods returns the first occurrence of the given element from the tuple.

```python
tup=(1,2,3,4,5,6,7,8,"Sagar")
tup.index(1)
```

```python
tup=(1,2,3,4,5,1,6,7,8,"Sagar",1)
tup.index(1,3,8)     #(1,start,end)
```

```python
len(tup)
```

```python
tup1=(1,2,3)
new_tuple=tup1+(4,5)
print(new_tuple)
```
