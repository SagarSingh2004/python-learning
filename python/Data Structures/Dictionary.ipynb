1) Keys are unique, immutable(like strings, numbers, tuples).

```python
dic1={"Name":"Sagar","Age":1,58:"sehal",58:"Neha"}
print(dic1)
type(dic1)
```

### Merge (|) Operator

1) Creates a new dictionary by merging dict1 and dict2.
2) Does not modify the original dictionaries.
3) If keys overlap, values from the right dictionary overwrite the left.

```python
dic1 = {"a": 1, "b": 2}
dic2 = {"b": 20, "c": 3}

dic3 = dic1 | dic2
print(dic3)  # {"a": 1, "b": 20, "c": 3}
print(dic1)  # {"a": 1, "b": 2} → unchanged
print(dic2)  # {"b": 20, "c": 3} → unchanged
```

### Difference between shallow and deep copy

```python
d1={"a":1,"b":[2,3]}
d2=d1.copy()
d2["a"]=10
d2["b"].append(4)   #This copy() is shallow copy if we change items in a nested mutable objects like list inside in dict it changes the dict also
print(d1)
print(d2)
```

```python
import copy as cp
d1={"a":1,"b":[2,3]}
d2=cp.deepcopy(d1)
d2["b"].append(4)   #This copy.deepcopy() independent of the original dict
print(d1)
print(d2)
```

### Sorted

To sort a dictionary by keys, you can create a new sorted dictionary.

```python
my_dict = {"b": 2, "c": 3, "a": 1}

# Sort by keys
sorted_dict = dict(sorted(my_dict.items()))
print(sorted_dict)
```

```python
my_dict = {"b": 2, "c": 3, "a": 1}

# Sort by keys
sorted_dict = dict(sorted(my_dict.items()))
print(sorted_dict)
```

```python
dic={"f":1,"a":2,"h":3}
new_dic=dict(sorted(dic.items()))
print(new_dic)
```

To sort by values, you need to tell sorted() to use the value as the key

```python
dic = {"f": 1, "a": 2, "h": 3}
# Sort by values
new_dic = dict(sorted(dic.items(), key=lambda item: item[1]))
print(new_dic)
# dic.items() → gives key-value pairs
# key=lambda item: item[1] → sorts based on value (second element of the tuple)
# dict() → converts the sorted pairs back into a dictionary
```
