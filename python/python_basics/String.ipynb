# Strings

In python, String is a sequence of characters enclosed between single and double quotation mark is considered as string.

```python
Name="Sagar"
Age='21'
print("My name is",Name,"and My age is",Age) 
```

Sometimes, we want to add single or double quotation marks in between our strings. we can use escape sequence or

```python
# So if we want to add single qoutation in between string so we make string with double quotation mark and vice versa.
print("Hello,'My name is sagar'")
print('Hello,"My name is sagar"')
```

### Multiple line string

If we want to add multiple lines in string so we have to enclosed between triple single or double quotation mark.

```python
# for ex:
A="""My name is sagar
i am 21 years old
i live in faridabad"""
print(A)
```

### Accessing characters of a string(Indexing)

In python, string is like an array of characters. we can access characters of string by using their index starts from 0 and the negative indexing starts from end.

```python
a="Sagar"
print(a[0])
print(a[1])
print(a[2])
print(a[3])
print(a[4])
print(a[-1])
print(a[-5])
```

### Loop through a String

For accessing all the characters line by line we can run a for loop

```python
a="Sagar"
for char in a:
    print(char)
```

### Length of a String

We can find the length of string by using len() function.

```python
a="My name is Sagar Singh"
print(len(a))
```

### String Slicing

Extract part of a string using [Start:End:Step]

```python
a="Alphabet"
print(a[0:5])
print(a[1:5])
print(a[:5])     # print(a[0:5])
print(a[:])      # print(a[0:len(a)])
print(a[0:-3])   # print(a[0:len(a)-3])
print(a[-8:])    # print(a[len(a)-8:len(a)])
print(a[::3])    # Give 0th index character and then the third one i.e. 0,3,6
```

### String Methods

In python, wehave built in methods to alter and modify the strings. The string is immutable, the methods create a new string don't change the existing one.

1) upper(): this method converts a string into upper case. For ex:

```python
a="sagar singh"
print(a.upper())
print(a)
```

2) lower(): this method converts a string into lower case. For ex:

```python
a="SAGAR SINGH"
print(a.lower())
print(a)
```

3) rstrip(): this method removes any trailing characters from right side.
   lstrip(): this method removes any leading characters from left side. For ex:

```python
a="!Sagar!!!!$$$$@@@"
print(a.rstrip("!$@"))
```

```python
a="Sagar    "      #remove space from right side
print(a.rstrip()) 
print(a)
```

```python
a="sagar\n\n"
print(a.rstrip())
print(a)
```

```python
a="!!!!Sagar!!!"
print(a.lstrip("!"))
```

4) strip(): this method removes characters from both left and right end of a string. For ex:

```python
a="\n\nSagar"
print(a.lstrip())
print(a)
```

```python
a="    Sagar     "
print(a.strip())
```

5) replace(): This method replaces all the occurence of a string with anothe string. For ex:

```python
a="Sagar Sing"
print(a.replace("Sing","Singh"))
```

6) split(): This method splits the string at the specified instance and return the string as list items.

```python
a="Sagar Singh Rawat"
print(a.split(" "))
```

7) capitalize(): This method turns only the first letter capital and rest of the small.

```python
a="saGar singh"
print(a.capitalize())
```

8) center(): This method alligns the string to the center as per the user parameters. For ex:

```python
a="Hello Friends"
print(a.center(50))
print(a.center(100))
```

9) count(): This method returns the number of times the given value has occurred within the given string. For ex:

```python
a="Sagar"
print(a.count("Sagar"))
print(a.count("a"))
```

10) endswith(): This method  checks if the string ends with a given value. if the valuse matches it gives true, else false

```python
a="Sagar Singh"
print(a.endswith("Singh"))
print(a.endswith("h"))
# we can also check with the start and end index positioning.
print(a.endswith("r",0,5))
```

11) find(): This method searches for the first occurence of the given value and returns the index where it is present. if value is not find in the string then it returns -1. 

```python
a="Sagar Singh"
print(a.find(" S"))
print(a.find("a"))
print(a.find("z"))
print(a.find("a",2))
```

12) index(): This method searches for the first occurence of the given value and returns the index where it is present. if value is not find in the string then it returns exception.

```python
a="Sagar Singh Rawat"
print(a.index("Singh"))
print(a.index("S",5))
print(a[::-1].index("t")) # firstly we reverse the string by negative slicing then using index function
```

13) isalnum(): This method returns True only if the entire string only consists of A-Z, a-z, 0-9, if any other characters or punctuation found it returns False.

```python
Username="Sagar123singh"
Email="itsofficialworkforss@mail.com"
print(Username.isalnum())
print(Email.isalnum())
```

14) isalpha(): This method returns Trueonly if the entire string only consist of A-Z, a-z, if any other character, notation, and numerical found it returns False.

```python
Username="SagarSingh"
User="Sagar Singh"
print(Username.isalpha())
print(User.isalpha())
```

15) islower(): This methods returns True if the all the characters of string are in lower case.

```python
a="sagar"
b="Singh"
print(a.islower())
print(b.islower())
```

16) isupper():This methods returns True if the all the characters of string are in lower case.

```python
a="SAGAR"
b="Singh"
print(a.isupper())
print(b.isupper())
```

17) isprintable(): This method returns True if all the values within the given string is printable,if not then returns False.

```python
a="My name is Sagar Singh"
b="My name is \nSagar Singh"
print(a.isprintable())
print(b.isprintable())
```

18) isspace(): This method returns True if the string contains ony white space either space or tab, else returning False.

```python
a="Sagar  Singh"
b="   "
print(a.isspace())
print(b.isspace())
```

19) istitle(): This method returs true if the first letter of each word of the string is capital, else it returns false.

```python
a="My Name Is Sagar"
print(a.istitle())
print(not a.istitle())
```

20) startswith(): This methods check if the string starts with a given value. if yes then True else False.

```python
a="Sagar Singh"
print(a.startswith("Sagar"))
print(a.startswith("H"))
# we can also check with the start and end index positioning.
print(a.startswith("a",1,5))
print(a.startswith("a",1))
```

21) swapcase(): This method changes the lower case character to upper case and vice versa.

```python
a="My Name is Sagar Singh"
print(a.swapcase())
```

22) title(): This method capitalize first letter of each word within the string. 

```python
a="my name is sagar singh"
print(a.title())
```

23) join(): This method combines elements of an iterable( list, tuple, set) into a single string.

```python
# "seperator".join(variable)
a=["Apple","Banana","Mango"]
print("-".join(a))
```

```python
# "seperator".join(variable)
a=["Apple","Banana","Mango"]
b="-".join(a)
print(b)
c=b.split("-")
print(c)
```

```python
Name="Sagar"
Age=22
print(f"My name is {Name} and my age is {Age}")
```

### Palindrome

```python
s = "madam"
print(s[0:5])   # 'madam' (positive slicing)
print(s[-5:])   # 'madam' (negative slicing)
```

```python
s[::]==s[::-1]
```
