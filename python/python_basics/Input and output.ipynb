# Input and Output in Python

### Output in python

We use the print() function to display output. For ex:

```python
print("Hello World")
```

We also print multiple strings at once seperated by commas. For ex:

```python
name="Sagar"
Age=21
print("My name is",name,"and My age is",Age)
```

### Input in python

We use input() function to take input from the user.
By default, input is taken as a string.

```python
name=input("Enter Your Name: ")
print("My name is",name)
```

```python
age=input("Enter Your Age-")
print(age)
type(age)
# input is always taken as string
```

Since input is always a string, we convert if needed. This is called Type Casting.

```python
age=int(input("Enter your age:"))
print(age)
type(age)
```

```python
Name=input("Enter your Name:")
Age=int(input("Enter your Age:"))
print("Hi", Name, "You will be", Age+1,"Next year.")
```
