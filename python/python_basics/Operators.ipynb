# Operators in python

Operators are symbols that perform operations on variables and values. Python have different types of operators for different operations.

1) Arithmetic operator: Used for mathematical claculations.

|Operator|Name|Example|Output|Comment|
|---|---|---|---|---|
|+|Addition|15+2|17|adds|
|-|Subtraction|15-2|13|subtract|
|*|Multiplication|15*2|30|multiplication|
|**|Exponential|15**2|225|power|
|/|Division|15/3|5.0|divide and output is Float|
|//|Floor Division|15//3|5|divide but output integer|
|%|Modulus|5%3|2|output is remainder|

```python
#Arithmetic operations
x=15+2
y=15-2
z=15*2
a=15**2
b=15/3
c=15//3
d=5%3
print(x)
print(y)
print(z)
print(a)
print(b)
print(c)
print(d)
```

2) Comparison operators: Used to compare values, result is True/False.

|Operator|Name|
|---|---|
|==|Equal|
|!=|Not equal|
|>|Greater than|
|<|Less than|
|>=|Greater than equal|
|<=|Less than equal|

```python
# Comparison Operator
a=10
b=20
print(a==b)
print(a!=b)
print(a>b)
print(a<b)
print(a>=b)
print(a<=b)
```

3) Logical Operator: Used to combine conditional statements.

|Operator|Name|Output|
|---|---|---|
|and|Logical and|Returns True if both conditions are True, otherwise False|
|or|Logical or|Returns True if at least one condition is True, returns Flase only if both are false|
|not|Logical not|If condition is True becomes False, If condition is False becomes True|

```python
# Logical operations
# with boolean Values
x=True
y=False
print(x and y)
print(x or y)
print(not x)
# with non boolean Values
# False values: 0, " ", empty list, empty dict, none, false
# Truthy values: everything else
print(10 and 20)
print(0 and 20)
print(10 or 20)
print(0 or 20)
print(not 10)
print(not 0)
```

5) Bitwise operators: works on binary bits. Bitwise or is |.

|Operator|Name|Conditions|
|---|---|---|
|&|Bitwise And|Compare each bits of two numbers, result is 1 only if both bits are 1, else 0|
|\||Bitwise Or|Compare each bits of two numbers, result is 1 if at least one bit is 1|
|^|Bitwise Xor|Exclusive Or, result is 1 if bits are different, 0 if they are same|
|~|Bitwise Not|inverts each bit, it also give negative numbers because of how integer is stored ~a=-(a+1)|
|<<|Left Shift|Shifts bits to the left by given positons, Each left shift=multiply by 2|
|>>|Right Shift|Shifts bits to the right by given positions, Each right shift=integer divide by 2|

```python
# Bitwise operations
a=5
b=6
print(a&b)
print(a|b)
print(a^b)
print(~10)
print(a<<1)
print(a<<2)
print(a>>1)
print(a>>2)
2^2
4^2
```

6) Membership operators: Check if a value exists in a sequence.

|Operator|Name|Output Condition|
|---|---|---|
|in|Membership|Returns true if value present in sequence|
|not in|Non-Membership|Returns true if valuse is not present in sequence|

```python
#Membership operation
fruits=["apple","banana","Mango"]
print("apple" in fruits)
print("sagar" in fruits)
text="Sagar"
print("H" not in text)
print("a" not in text)
```

7) Identity operators:  Check if two variables point to the same object in memory.

|Operator|Name|condition|
|---|---|---|
|is|identity|Returns True if both variables refer to the same object(same memory location)|
|is not|Not identity|Returns True if both variables do not refer to the same object.|

```python
# Identity operators
a=[1,2,3]
b=a
c=[1,2,3]
print(a is b)
print(b is a)
print(a is c)
print(c is not a)
print(a is not b)
```
