# Escape Sequence

Escape sequences are used to represent special characters in a string.

In python, Escape sequence starts with a backslash(\) followed by a special character.

Common escape sequence includes:
|Escape Sequence|Name|
|---|---|
|\n|New line|
|\t|Tab|
|\\\\ |Backslash|
|\\'|Single quote|
|\\"|Double quote|

```python
print("My name is \t Sagar")
```

```python
print("My name is \nSagar singh")
```

```python
print("My name is \\ Sagar singh")
```

```python
print("my name is \ sagar")
```

```python
print("My name is \"Sagar Singh\"")
```
