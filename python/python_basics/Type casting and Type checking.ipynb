# Type Casting and Type Checking

### Type casting (Type Conversion)

Converting one data type into another.

There are two type of Type casting:
1) Implicit Type casting: Done automatically by python.
2) Explicit Type casting: Done manually using Functions.

```python
# Rules for implicit: it always promotes smaller precise types into larger precise types.
# int + float : float
# float + complex : complex
# int + complex : complex
```

```python
# Implicit type casting
x=10
y=2.5
z=x+y
print(z)
```

```python
# # Builtin functions to manually convert data types.
# # int(): converts to integer
# float(): converts to float
# complex(): convert to complex number
# str(): convert to string
# bool(): convert to boolean( 0,"",None to False else True)
# list(): convert to list (from string not numeric data type)
# tuple(): convert to tuple (from string not numeric data type)
# set(): convert to set
# dict(): convert to dictionary(only from key value pairs)
```

```python
# Explicit type casting
a=int(3.9)
b=float(10)
c=complex(5)
x=list("123") #'int' object is not iterable
y=tuple("123") #'int' object is not iterable
print(a,b,c,x,y)
```

```python
pairs=[("Apple",250),("Banana",100)]
print(dict(pairs))
```

### Type Checking

We use type() function to check data type.

```python
x="sagar"
y=[1,2,"Sagar"]
z=1
type(x),type(y),type(z)
```

Check data type using isinstance() function. Gives true or false as output.

```python
x=25
print(isinstance(x,int))
print(isinstance(x, str))
```
