Q1. Create the first numpy array of 1-D with values (1,2,3,0). 
Create a second numpy array of 1-D with values (4,1,0, 
6). Perform following simple arithmetic operations on 
these two numpy arrays: 
• Addition 
• Subtraction 
• Multiplication 
• Division

Answer:

```python
import numpy as np

a=np.array([1,2,3,0])          #first numpy array
b=np.array([4,1,0,6])          #Second numpy array

#arithematic operations on a and b
print("a:",a)
print("b:",b)
print("\nAddition of a and b:\n",a+b)         #addition
print("\nSubtraction of a and b:\n",a-b)         #subtraction
print("\nMutliplication of a and b:\n",a*b)         #Multiplication
print("\nDivision of a and b:\n",a/b)         #Division
```

Q3. Create a 3-D array of 27 random elements using random 
function of shape 3*3*3. Now pull the following 
elements using indexing/slicing 
• Third column of second 2-D array 
• First and third row of first 2-D array 
• Intersection of first row and third column in 
third 2-D array

Answer:

```python
import numpy as np

a=np.random.rand(3,3,3)      # 27 random elements using random function of shape 3*3*3

print("\nOriginal array:\n",a)                                                                                # print array a
print("Third column of second 2-D array:",a[1,:,2])                                     # Third column of second 2-D array
print("First and third row of first 2-D array:",a[0,[0,2]])                             # First and third row of first 2-D array
print("Intersection of first row and third column in third 2-D array:",a[2,0,2])        # Intersection of first row and third column in third 2-D array
```

Q4. Create a 2-D numpy array of random integers between 
1 to 100 with shape 5*4. Now perform following 
operations: 
• Print only the even numbers from this array 
• Print only the odd numbers from this array 
• Print row wise sum of all elements 
• Print column wise sum of all elements 
• Convert this 2-D array into 3-D array of shape 
2*2*5

Answer:

```python
import numpy as np

a=np.random.randint(1,101,(5,4))              # 2D array of random integers between 1 to 100 of shape(5,4)

print("\nOriginal array:\n",a)                                      # array a
print("\nEven numbers from array a:\n",a[a%2==0])                              # Print only the even numbers from this array
print("\nOdd numbers from array a:\n",a[a%2==1])                              # Print only the odd numbers from this array
print("\nRow wise sum:\n",np.sum(a,axis=1))                       # Print row wise sum of all elements
print("\nColumn wise sum wise sum:\n",np.sum(a,axis=0))                       # Print column wise sum of all elements

b=np.reshape(a,(2,2,5))                       # reshaping array a into (2,2,5) as b
print("\nNew shape of array a(2,2,5):\n",b)
```

Q5. Create a 2-D numpy array of random integers between 
1 to 100 with shape 5*4. Now perform following 
operations: 
• Swap first row with third row 
• Swap second column with fourth column. 
• Replace the values less than 50 with zero. 
• Convert this 2D array into 1-D array

Answer:

```python
import numpy as np

a=np.random.randint(1,101,(5,4))              # 2D array of random integers between 1 to 100 of shape(5,4)

print("Original array:\n",a)
a[[0,2]]=a[[2,0]]
print("\nAfter swapping first and third row:\n",a)
a[:,[1,3]]=a[:,[3,1]]
print("\nAfter swapping second and fourth column:\n",a)
b=np.where(a<50,0,a)
print("\nAfter replacing values < 50 with 0:\n",b)
a=a.flatten()
print("\nFlattened 1D array:\n", a)
```

Q6. Explain the concept of stacking in numpy with examples.

differences between them.

Answer:

| Stacking| Concatenation|
|----|----|
|Adds a new axis when joining arrays (e.g., from 2D → 3D).| Joins arrays along an existing axis only.|
| Uses functions like `np.stack()`, `np.hstack()`, `np.vstack()`, `np.dstack()`.| Uses `np.concatenate()` with `axis` parameter.|
